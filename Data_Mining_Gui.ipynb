```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
from tkinter import *
import tkinter.font as tkFont


def run():
    area = area_storage.get()
    bedrooms = bedrooms_storage.get()
    age = age_storage.get()
    floor = floor_storage.get()
    value = clicked.get()
    
    if value=="dhanmondi":
        sum = model.predict([[bedrooms,age,floor,area,0,1]])
        
    elif value=="banani":
        sum = model.predict([[bedrooms,age,floor,area,1,0]])
        
    elif value=="zigatola":
        sum = model.predict([[bedrooms,age,floor,area,0,0]])
    
    new_text = Label(text = sum, fg = "green", bg = "white", font=fontStyle_result)
    new_text.place(x = 120, y = 400)


df = pd.read_csv("hard_mining.csv")
df

import math
median_bedrooms = math.floor(df.bedrooms.median())
median_bedrooms

df.bedrooms = df.bedrooms.fillna(df.bedrooms.median())
df

dummies = pd.get_dummies(df.location)
merged = pd.concat([df,dummies],axis='columns')
final = merged.drop(['location'], axis='columns')
final = final.drop(['zigatola'], axis='columns')

x = final.drop('price', axis='columns')
y = final.price

from sklearn.linear_model import LinearRegression
model = LinearRegression()

model.fit(x,y)
    
screen = Tk()
screen.title("House Price Detection System")
screen.geometry("500x500")

fontStyle = tkFont.Font(family="Lucida Grande", size=20)
fontStyle_result = tkFont.Font(family="Lucida Grande", size=20)
welcome_text = Label(text = "House Price Detection System", fg = "blue", bg = "violet", font=fontStyle)
welcome_text.pack()

area_text = Label(text = "Enter Area Size: ")
area_text.place(x=120,y=80)

area_storage = DoubleVar()
area = Entry(textvariable = area_storage)
area.place(x=300,y=80)


bedrooms_text = Label(text = "Enter Number of Bedrooms: ")
bedrooms_text.place(x=120,y=120)

bedrooms_storage = IntVar()
bedrooms = Entry(textvariable = bedrooms_storage)
bedrooms.place(x=300,y=120)


age_text = Label(text = "Enter House Age: ")
age_text.place(x=120,y=160)

age_storage = DoubleVar()
age = Entry(textvariable = age_storage)
age.place(x=300,y=160)


floor_text = Label(text = "Enter Floor Number: ")
floor_text.place(x=120,y=200)

floor_storage = IntVar()
floor = Entry(textvariable = floor_storage)
floor.place(x=300,y=200)

options = [
    "dhanmondi",
    "banani",
    "zigatola"
]

clicked = StringVar()
clicked.set(options[0])

drop_text = Label(text = "Choose Any Location: ")
drop_text.place(x=120,y=240)

drop = OptionMenu(screen, clicked, *options)
drop.place(x=300,y=240)

click = Button(text = "Calculate Price", fg = "red", bg = "yellow", command = run)
click.place(x=200, y=320)

mainloop()
```
