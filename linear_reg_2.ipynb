```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("final_format.csv")
df
```

```python
import math
median_bedrooms = math.floor(df.bedrooms.median())
median_bedrooms
```

```python
df.bedrooms = df.bedrooms.fillna(df.bedrooms.median())
df
```
