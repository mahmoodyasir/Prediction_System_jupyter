```python
d=int(input('How many Vertex the dataset have\n'))
c=int(input('How many Centroid do you have\n'))

dataset=[]
for i in range(0,d):
    print(f"Enter your dataset Vertex {i+1}st data\t\t")
    x=input('X=')
    y=input('Y=')
    x_y=[x,y]
    dataset.append(x_y)

Centroid=[]
for i in range(0,c):
    print(f"Enter your cluster {i+1}st data\t\t")
    x=input('X=')
    y=input('Y=')
    x_y=[x,y]
    Centroid.append(x_y)

import math
fi=[]
for l in range(0,c):
    li=[]
    for k in range(0,d):
        res=math.sqrt(((float(dataset[k][0])-float(Centroid[l][0]))**2)+((float(dataset[k][1])-float(Centroid[l][1]))**2))
        li.append(res)
    fi.append(li)

import pandas as pd
df=pd.DataFrame(fi)
df
```
